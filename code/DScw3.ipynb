```python
used_cols = ['Calories', 'Calories from Fat', 'Total Fat', 'Cholesterol', 'Sugars']
# highest nurition
max_idxs = [menu_data[col].idxmax() for col in used_cols]
for col, max_idx in zip(used_cols, max_idxs):
    print('{} highest item：{}'.format(col, menu_data.iloc[max_idx]['Item']))
```

```python
min_idxs = [menu_data[col].idxmax() for col in used_cols]
for col, min_idx in zip(used_cols, min_idxs):
    print('{} lowest item：{}'.format(col, menu_data.iloc[min_idx]['Item']))
```

```python
import seaborn as sns
cols = ['Calories','Cholesterol','Trans Fat','Sugars','Dietary Fiber']
cm = np.corrcoef(menu_data[cols].values.T)
sns.set(font_scale = 1.5)
hm = sns.heatmap(cm,cbar = True, annot = True,square = True, fmt = '.2f', annot_kws = {'size':15}, yticklabels = cols, xticklabels = cols)
```

```python
import matplotlib.pyplot as plt
f,ax=plt.subplots(figsize=(18,11))
sns.heatmap(menu_data.corr(),annot=True,linewidths=5,fmt='.1f',ax=ax)
```

```python
cat_grouped = menu_data.groupby('Category')
print('menu ：')
print(cat_grouped.size().sort_values(ascending=False))
```

```python
print('Distribution of nutrients in menu types：')

used_cols = ['Calories', 'Calories from Fat', 'Total Fat', 'Cholesterol', 'Sugars']
print(cat_grouped[used_cols].mean())
```

```python
max_cats = [cat_grouped[col].mean().idxmax() for col in used_cols]
for col, cat in zip(used_cols, max_cats):
    print('{} highest of menu：{}'.format(col, cat))
```

```python
min_cats = [cat_grouped[col].mean().idxmax() for col in used_cols]
for col, cat in zip(used_cols, min_cats):
    print('{} lowest of menu：{}'.format(col, cat))
```
